```python
col_cles_camembert = [
    "candidats contactés par AJC",
    "Candidats profil jugé KO par AJC",
    " NRP",
    "candidats écartés par AJC",
    "CANDIDATURES ENVOYEES"
]
```

```python
colonnes_camembert = [col for col in dff.columns if any(mot in col for mot in col_cles_camembert)]
```

```python
valeurs = dff.iloc[30][colonnes_camembert].fillna(0).values
labels = colonnes_camembert
```

```python
valeurs
```

```python
dff.iloc[30][colonnes_camembert]
```

```python
# Tracer le camembert
plt.figure(figsize=(8, 8))
plt.pie(valeurs, labels=labels, autopct='%1.1f%%', startangle=140)
plt.title("Camembert dynamique")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Pour le style 3D

# Définir les colonnes pour le graphique en barres
col_cle_bar = [
     " CV TRAITES",
    "Candidats profil jugé KO par AJC",
    "candidats contactés par AJC",
    "candidats écartés par AJC",
    "CANDIDATURES ENVOYEES"
]

# Extraire les valeurs et les étiquettes
valeurs = dff.iloc[30][col_cle_bar].fillna(0).values
labels = col_cle_bar

# Créer une figure 3D
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111, projection='3d')

# Positions des barres sur l'axe x
x_pos = range(len(labels))

# Créer les barres 3D
ax.bar(x_pos, valeurs, zs=0, zdir='y', color='skyblue', alpha=0.8)

# Ajouter les valeurs au-dessus des barres
for i, v in enumerate(valeurs):
    ax.text(i, 0, v + 10, str(int(v)), ha='center', va='bottom', fontsize=10)

# Définir les étiquettes des axes
ax.set_xticks(x_pos)
ax.set_xticklabels(labels, rotation=45, ha='right')
ax.set_yticks([])  # Cacher l'axe y (inutile pour un graphique 3D de ce type)
ax.set_zlabel('Valeurs')

# Ajuster la disposition
plt.tight_layout()

# Afficher le graphique
plt.show()
```
